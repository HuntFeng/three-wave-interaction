```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
files = os.listdir("data/a0")
files = sorted(files,key=lambda name: int(name[:-4]))

n = 331
frame = 100*n

x = np.load("data/x.npy")
t = np.load("data/t.npy")
a0 = np.load(f"data/a0/{files[n]}")
a1 = np.load(f"data/a1/{files[n]}")
a2 = np.load(f"data/a2/{files[n]}")

plt.plot(x,np.sum(a0**2,axis=1), label="pump, $|a_0|^2$")
plt.plot(x,np.sum(a1**2,axis=1), label="seed, $|a_1|^2$")
plt.plot(x,np.sum(a2**2,axis=1), label="plasma, $|a_2|^2$")
plt.legend()
plt.title(f"$t={t[frame]:.6f} ps$")
x0 = 10
v = 298
plt.ylim([0,0.04])
plt.xlim([x0+v*t[frame]-250, x0+v*t[frame]+250])
plt.xlabel("$x(\mu m)$")
plt.ylabel("normalized energy")
```
